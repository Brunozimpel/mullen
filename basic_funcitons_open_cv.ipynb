# Playing around with open cv

```python
import cv2
import numpy as np
import utils as u
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reading the image

```python
img = cv2.imread(u.get_sample(0))
u.plot_single_image(img)
```

## Color space

### Inverting

```python
inv_img = 255 - img
u.plot_single_image(inv_img)
```

### Grayscale

```python
gray_image = cv2.cvtColor( cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), cv2.COLOR_GRAY2RGB)
u.plot_single_image(gray_image)
```

### Binarizing

```python
bin_img = (gray_image >= 128)*255
u.plot_single_image(bin_img)
```

## Noise removal

### Gaussian Blur

```python
noiseless_img = cv2.GaussianBlur(gray_image,(3,3),0)
u.plot_side_by_side((img, noiseless_img))
```

## Edge detecting

### Canny

```python
edge_img = cv2.Canny(img,100,200)
u.plot_single_image(edge_img)
```
