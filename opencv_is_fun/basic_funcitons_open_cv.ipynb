### Playing around with open cv

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_sample(num, img_dir='../samples/img/'):
    """Quick alias to get the image in the correct path"""
    import os.path
    file_path = '%ss%i.jpg' % (img_dir, num)
    if os.path.exists(file_path):
        return file_path
    else:
        raise ValueError('File not found')             
```

```python
get_sample(1)
```

Reading the image

```python
img = cv2.imread(get_sample(0))
plt.imshow(img)
pass
```

# Inverting

```python
inv_img = 255 - img
plt.imshow(inv_img)
pass
```

Grayscale

```python
gray_image = cv2.cvtColor( cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), cv2.COLOR_GRAY2RGB)
plt.imshow(gray_image)
pass
```

Binarizing

```python
bin_img = (gray_image >= 128)*255
plt.imshow(bin_img)
pass
```
